```python
# y = e^(x-2) + 2
# y = (x-2)^2
import numpy as np
x = [i for i in np.arange(0,11,0.01)]
y1 = [np.exp((i-2)) for i in x]
y2 = [(i-2)**2 for i in x]
plt.plot(x,y1,'b',label='y1 = e^(x-2) + 2')
plt.plot(x,y2,'r',label='y2 = (x-2)^2')
plt.xlabel("height"); plt.ylabel("weight"); plt.title("relation of height and weight")
plt.legend(loc='upper left')
plt.axis([-2,11,-20,20])
```
